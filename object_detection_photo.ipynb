```python
import cvlib as cv
from cvlib.object_detection import draw_bbox
import cv2
from google.colab.patches import cv2_imshow
```

```python
!wget https://www.pinterest.com/pin/282178732892452221/
!wget https://www.pinterest.com/pin/303922674859945743/
```

```python
image = cv2.imread('object_detection.jpg')
```

```python
(bbox, label, conf) = cv.detect_common_objects(image)
print(bbox, label, conf)
```

```python
out = draw_bbox(image, bbox, label, conf)
```

```python
cv2_imshow(out)
```

```python
cv2.imwrite('object_detection_output.jpg', out)
```
